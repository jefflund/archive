```python
import io
import numpy as np
import numpy.random as rng
from skimage import draw
from PIL.Image import fromarray
from IPython.display import display, Image
```

```python
HEIGHT, WIDTH = 288, 512
NUM_LINES = 50
```

```python
image = np.full((HEIGHT, WIDTH, 3), 0, dtype=np.uint8)

for _ in range(50):
    r0, r1, r2 = rng.randint(HEIGHT, size=3)
    c0, c1, c2 = rng.randint(WIDTH, size=3)
    rr, cc = draw.bezier_curve(r0, c0, r1, c1, r2, c2, 1.5)
    color = rng.random(3) * 255
   
    for i in range(10):
        draw.set_color(image, (rr+i, cc), color)
```

```python
buf = io.BytesIO()
fromarray(image).save(buf, 'png')
display(Image(buf.getvalue()))
```
